```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file in Colab
file_path = "/content/Liju.xlsx"  # Ensure the file path matches where you uploaded it in Colab
sheets = ["PD0.5", "PD1", "PD1.5", "PD2", "PD2.5", "PD3"]

# Load all sheets into a dictionary
data = {sheet: pd.read_excel(file_path, sheet_name=sheet) for sheet in sheets}

# Define colors for differentiation in plots
colors = ["b", "g", "r", "c", "m", "y"]

# Nyquist Plot (Z' vs Z'')
plt.figure(figsize=(10, 6))
for idx, sheet in enumerate(sheets):
    plt.plot(data[sheet]["Z'"], data[sheet]["Z''"], marker='o', linestyle='', color=colors[idx], label=sheet)
plt.xlabel("Z' (Real Impedance)")
plt.ylabel("Z'' (Imaginary Impedance)")
plt.title("Nyquist Plot")
plt.legend()
plt.grid(True)
plt.show()

# Cole-Cole Plot (e' vs e'')
plt.figure(figsize=(10, 6))
for idx, sheet in enumerate(sheets):
    plt.plot(data[sheet]["e'"], data[sheet]["e''"], marker='o', linestyle='', color=colors[idx], label=sheet)
plt.xlabel("e' (Dielectric Real)")
plt.ylabel("e'' (Dielectric Imaginary)")
plt.title("Cole-Cole Plot")
plt.legend()
plt.grid(True)
plt.show()

# Frequency Dependence Plot for Impedance (f vs Zm and Zp)
plt.figure(figsize=(10, 6))
for idx, sheet in enumerate(sheets):
    plt.plot(data[sheet]["f"], data[sheet]["Zm"], color=colors[idx], label=f"Zm - {sheet}")
    plt.plot(data[sheet]["f"], data[sheet]["Zp"], linestyle='--', color=colors[idx], label=f"Zp - {sheet}")
plt.xscale("log")
plt.xlabel("Frequency (f)")
plt.ylabel("Impedance")
plt.title("Frequency Dependence of Impedance")
plt.legend()
plt.grid(True)
plt.show()

# Frequency Dependence Plot for Dielectric Values (f vs e' and e'')
plt.figure(figsize=(10, 6))
for idx, sheet in enumerate(sheets):
    plt.plot(data[sheet]["f"], data[sheet]["e'"], color=colors[idx], label=f"e' - {sheet}")
    plt.plot(data[sheet]["f"], data[sheet]["e''"], linestyle='--', color=colors[idx], label=f"e'' - {sheet}")
plt.xscale("log")
plt.xlabel("Frequency (f)")
plt.ylabel("Dielectric Values")
plt.title("Frequency Dependence of Dielectric Values")
plt.legend()
plt.grid(True)
plt.show()

# Frequency Dependence Plot for Modulus Values (f vs M' and M'')
plt.figure(figsize=(10, 6))
for idx, sheet in enumerate(sheets):
    plt.plot(data[sheet]["f"], data[sheet]["M'"], color=colors[idx], label=f"M' - {sheet}")
    plt.plot(data[sheet]["f"], data[sheet]["M''"], linestyle='--', color=colors[idx], label=f"M'' - {sheet}")
plt.xscale("log")
plt.xlabel("Frequency (f)")
plt.ylabel("Modulus")
plt.title("Frequency Dependence of Modulus Values")
plt.legend()
plt.grid(True)
plt.show()

# Frequency Dependence Plot for Dielectric Loss (f vs tan d)
plt.figure(figsize=(10, 6))
for idx, sheet in enumerate(sheets):
    plt.plot(data[sheet]["f"], data[sheet]["tan d"], color=colors[idx], label=f"tan d - {sheet}")
plt.xscale("log")
plt.xlabel("Frequency (f)")
plt.ylabel("Dielectric Loss (tan d)")
plt.title("Frequency Dependence of Dielectric Loss")
plt.legend()
plt.grid(True)
plt.show()
```
